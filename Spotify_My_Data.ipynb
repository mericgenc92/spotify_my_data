```python
import pandas as pd
import numpy as np
```

```python
df_stream = pd.read_json("StreamingHistory0.json")
df_stream.head()
```

```python
df_stream['UniqueID'] = df_stream['artistName'] + ":" + df_stream['trackName']

df_stream.head()
```

```python
import json
from pandas.io.json import json_normalize

with open("YourLibrary.json") as json_data:
    data = json.load(json_data)

df_library = pd.DataFrame(data["tracks"])
df_library.head()
```

```python
df_library["UniqueID"] = df_library["artist"] + ":" + df_library["track"]
df_library.head()
```

```python
new = df_library["uri"].str.split(":", expand = True)
new.head()
```

```python
df_library["track_uri"] = new[2]
df_library.head()
```

```python
df_tableau=df_stream.copy()
df_tableau.head()
```

```python
df_tableau["In Library"]=""
```

```python
df_tableau['In Library'] = np.where(df_tableau["UniqueID"].isin(df_library["UniqueID"]).tolist(), 1, 0)
```

```python
df_tableau = pd.merge(df_tableau, df_library[["album","UniqueID","track_uri"]], how="left", on=["UniqueID"])
```

```python
df_tableau.head()
```

```python
import spotipy
CLIENT_ID = 'Your_Client_ID'
CLIENT_SECRET = 'Your_Client_Secret'
```

```python
auth_manager = spotipy.SpotifyClientCredentials(client_id=CLIENT_ID,
                                        client_secret=CLIENT_SECRET,)

sp = spotipy.Spotify(client_credentials_manager=auth_manager)
```

```python
genres = []
for a in df_tableau['artistName']:
    genres.append(sp.search(a, type='artist')['artists']['items'][0]['genres'])
df_tableau['genres'] = genres
```

```python
df_tableau.head()
```

```python
df_tableau_expanded = df_tableau.explode('genres')
df_tableau_expanded.index = range(0,len(df_tableau_expanded))
df_tableau_expanded.tail()
```

```python
dict_genre = {}
track_uris = df_library['track_uri'].to_list()

for t_uri in track_uris:
    dict_genre[t_uri] = {'artist_uri': "", "genres":[]}
    a_uri = sp.track(t_uri)["artists"][0]["uri"].split(":")[2]
    a_name = sp.track(t_uri)['artists'][0]['name']
    t_name = sp.track(t_uri)['name']
    dict_genre[t_uri]['artist_uri'] = a_uri
    dict_genre[t_uri]['artist_name'] = a_name
    dict_genre[t_uri]['track_name'] = t_name

    a_genres = sp.artist(a_uri)['genres']
    dict_genre[t_uri]['genres'] = a_genres
```

```python
df_genre = pd.DataFrame.from_dict(dict_genre, orient='index')
df_genre.insert(0, 'track_uri', df_genre.index)
df_genre.reset_index(inplace=True, drop=True)

df_genre.head()
```

```python
df_genre_expanded = df_genre.explode('genres')
df_genre_expanded.index = range(0,len(df_genre_expanded))
df_genre_expanded.head()
```

```python
df_tableau.to_csv('tables/MySpotifyDataTable.csv')
df_tableau_expanded.to_csv('tables/MySpotifyExpandedTable.csv')
df_genre_expanded.to_csv('tables/GenresExpandedTable.csv')

print('done')
```
